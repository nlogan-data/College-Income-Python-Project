# 02 Wrangling

## 02.01 Drop Columns

```python
#Drop columns part 1
drop_list = ['opeid', 'opeid6', 'sat_verbal_25_percentile'
             , 'sat_verbal_75_percentile', 'sat_math_25_percentile'
             , 'sat_math_75_percentile', 'sat_verbal_median'
             , 'sat_math_median', 'sat_equivalent_avg_all']

ced = ced.drop(columns = drop_list)
ced.head()
```

```python
#Drop columns part 2
#Drop last N columns (all fraction_majors_X columns)
N = 39
ced = ced.iloc[:, :-N]
ced.head()
```

# 03 Cleaning

## 03.01 Nulls and Data Types

```python
ced.info()
```

#### Due to the nature of the variables, we can infer that all null values represent missing values--no other significance. There are 3645 entries, so any variable with less than 2916 non-null values (729 or more nulls) would have 20% or more nulls. Those variables should be dropped. Unfortunately, this throws out all the demographic completion rate variables.

```python
#Drop columns with >= 20% nulls (729 nulls)
ced = ced[[col for col in ced if ced[col].isnull().sum() <= 729]]
ced.head()
```

```python
ced.info()
```

```python
#Change unitid from int to string
ced['unitid'] = ced['unitid'].astype('str')
ced['unitid'].dtype
```

```python
#Check for mixed-type data
for col in ced.columns.tolist():
    weird = (ced[[col]].applymap(type) 
             != ced[[col]].iloc[0].apply(type)).any(axis = 1)
    if len(ced[weird]) > 0:
        print(col)
```

#### No columns with mixed-type data
